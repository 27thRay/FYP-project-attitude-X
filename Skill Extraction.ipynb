```python
import vercel_ai

client = vercel_ai.Client()

prompt = f'''
Resume:
{resume}'''

prompt_template = f'''
Only extract the name, technical skills, soft skills and language,
extract nothing else, do not write additional notes.
{prompt}'''

while True:
  try:
    result = ""
    for chunk in client.generate("replicate:replicate/llama-2-70b-chat", prompt_template, params = {
      "replicate:replicate/llama-2-70b-chat": {
        "temperature": 0.75,
        "maximumLength": 1000,
        "topP": 1,
        "repetitionPenalty": 1
    }}):
      result += chunk
    break
  except:
    pass
```

```python
import pandas as pd
import os

test = result.replace('\n*',',')
test = test.replace("\n,",'')
test = test.replace("\n\n",'\n')
test = test.split('\n')
outputdictionary = {}
for i in test:
    templist = i.split(':')
    print(templist)
    if len(templist[1].split(',')) > 1:
        stringifylist = templist[1].split(',')
        newlist = []
        for string in stringifylist:
            newstring = f'{string.strip()}'
            newlist.append(newstring)
        templist[1] = newlist
        outputdictionary[templist[0]] = templist[1]
    else:
        outputdictionary[templist[0]] = templist[1].strip()

if 'Outputs.csv' not in os.listdir('data'):
    df = pd.DataFrame(columns=['Data Dictionary'])
else: 
    df = pd.read_csv('data/Outputs.csv')
new_row = pd.DataFrame([f'{outputdictionary}'],columns=['Data Dictionary'])
df = pd.concat([df, new_row],ignore_index=True,axis=0)
df.to_csv('data/Outputs.csv',index=False)
print(df)
```
