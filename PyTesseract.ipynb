```python
# Download Windows installer from https://github.com/UB-Mannheim/tesseract/wiki and check into the directory of the tesseract executable
import pytesseract
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
```

```python
import os
from PIL import Image
from pdf2image import convert_from_path

filePath = "./sample-data/CV+Bertrand+Comte+-+June++2023+-+SWITZERLAND (1).pdf"
doc = convert_from_path(filePath)
path, fileName = os.path.split(filePath)
fileBaseName, fileExtension = os.path.splitext(fileName)

for page_number, page_data in enumerate(doc):
    txt = pytesseract.image_to_string(page_data).encode("utf-8")
    print(f"Page #{page_number}\n")
    txt = str(txt).replace(r'\n','\n')
    print(f"{txt}")
```
